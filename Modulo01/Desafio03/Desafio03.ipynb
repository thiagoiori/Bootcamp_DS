## Desafio 06
Plotar o gráfico dos custos apenas dos estados da região sudeste e verificar se os picos de 2013/Fev teve comportamento similar em todos os demais estados da região

```python
condition = [(ordenado_por_gasto.index.str[:1] == "1"),
             (ordenado_por_gasto.index.str[:1] == "2"),
             (ordenado_por_gasto.index.str[:1] == "3"),
             (ordenado_por_gasto.index.str[:1] == "4"),
             (ordenado_por_gasto.index.str[:1] == "5")]

values = ["Norte", "Nordeste", "Sudeste", "Sul", "Centro-Oeste"]

ordenado_por_gasto["Região"] = np.select(condition, values)
ordenado_por_gasto_SE = ordenado_por_gasto.loc[ordenado_por_gasto["Região"] == "Sudeste"]

ordenado_por_gasto_SE = ordenado_por_gasto_SE[colunas_interessadas]
ax = ordenado_por_gasto_SE.T.plot(figsize=(12,6))
ax.legend(loc="right", bbox_to_anchor=(.75, .2, 0.5, 0.5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos na saúde da Região SE")
plt.grid(axis="y")
plt.ylabel("Gasto em R$")
plt.show()
```
