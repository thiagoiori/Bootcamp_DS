Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

# Desafio 01

```python
ax = usaveis.T.plot(figsize=(12, 9))
ax.legend(loc="right", bbox_to_anchor=(.8, 0.2, 0.5, 0.5))
plt.show()
```

# Desafio 02

```python
estados = ["31 Minas Gerais", "13 Amazonas" ,"51 Mato Grosso", "24 Rio Grande do Norte", "43 Rio Grande do Sul"]
ax = usaveis.loc[estados].T.plot(figsize=(12, 9))
ax.legend(loc="right", bbox_to_anchor=(.8, 0.2, 0.5, 0.5))
plt.show()
```
